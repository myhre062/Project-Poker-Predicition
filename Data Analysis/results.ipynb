```python
# Load Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import pathlib as Path
```

```python
# Read the Results CSV in as a Pandas DataFrame
results_df = pd.read_csv("Model Performance Worksheet.csv")


# Review the DataFrame
results_df
```

```python
# Create a bar graph to show Accuracy results
plt.figure(figsize=(10, 6))

# Sort the DataFrame by "Best Accuracy Performance" in descending order
results_df_sorted = results_df.sort_values(by="Best Accuracy Performance", ascending=False)

plt.bar(results_df_sorted["Model Type"], results_df_sorted["Best Accuracy Performance"], color='mediumseagreen')


plt.xlabel("ML Model")
plt.ylabel("Accuracy %")
plt.title("Accuracy Performance of Machine Learning Models")
plt.xticks(results_df_sorted["Model Type"], rotation=45)
plt.show()
```
